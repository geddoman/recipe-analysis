```python
import pandas as pd
import matplotlib.pyplot as plt
from recipe_text_preprocessor import text_prepro
from nltk import FreqDist
from gensim.models import LdaModel,Doc2Vec,CoherenceModel
from gensim.corpora import Dictionary
import pyLDAvis.gensim
```

```python
recipe_file = 'full_recipes.csv'
df = pd.read_csv(recipe_file,index_col=0)
```

```python
df['method_cln'] = df['method'].apply(text_prepro)
```

```python
word_list = []

for item in df['method_cln'].values:
    for word in item:
        word_list.append(word)

freq_dist = FreqDist(word_list)
```

```python
plt.figure(figsize=(20,10))
freq_dist.plot(50)
```

```python
recipe_id2word = Dictionary(documents=df['method_cln'].values)
recipe_corpus = [recipe_id2word.doc2bow(recipe) for recipe in df['method_cln'].values]
```

```python
recipe_lda = LdaModel(corpus=recipe_corpus,id2word=recipe_id2word,num_topics=15)
```

```python
recipe_lda_vis = pyLDAvis.gensim.prepare(recipe_lda,recipe_corpus,recipe_id2word)
```

```python
pyLDAvis.display(recipe_lda_vis)
```

```python
from gensim.models import CoherenceModel
```

```python
ch_recipes = CoherenceModel(recipe_lda,corpus=recipe_corpus,texts=df['method_cln'].values,coherence='c_v')
```

```python
ch_recipes.get_coherence()
```
